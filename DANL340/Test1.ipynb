```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
SAMHSA_DC=pd.read_csv("../DANL340/SAMHSA D.C.csv")
SAMHSA_Maryland=pd.read_csv("../DANL340/SAMHSA Maryland.csv")
SAMHSA_Virginia=pd.read_csv("../DANL340/SAMHSA Virginia.csv")
DC_Crime=pd.read_csv("../DANL340/dc-crimes-search-results.csv")
Marlyand_Crime=pd.read_csv("../DANL340/Violent_Crime___Property_Crime_by_County__1975_to_Present.csv")
ACS_DC=pd.read_csv("../DANL340/ACSDT5Y2021_DC_pop.csv")
ACS_Maryland=pd.read_csv("../DANL340/ACSDT5Y2021_MD_pop.csv")
```

```python
DC_crime_rate=pd.concat([SAMHSA_DC,DC_Crime,ACS_DC], axis=1)
DC_crime_rate.head()
```

```python
DC_crime_rate_cat=DC_crime_rate.select_dtypes(object)
DC_crime_rate_num=DC_crime_rate.select_dtypes(np.number)
```

```python
DC_crime_rate_cat.isnull().sum()
```

```python
DC_crime_rate_cat.name1.fillna(DC_crime_rate_cat.name1.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.name2.fillna(DC_crime_rate_cat.name2.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.street1.fillna(DC_crime_rate_cat.street1.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.street2.fillna(DC_crime_rate_cat.street2.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.sector.fillna(DC_crime_rate_cat.sector.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.city.fillna(DC_crime_rate_cat.city.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.state.fillna(DC_crime_rate_cat.state.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.county.fillna(DC_crime_rate_cat.county.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.intake_prompt.fillna(DC_crime_rate_cat.intake_prompt.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.type_facility.fillna(DC_crime_rate_cat.type_facility.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.phone.fillna(DC_crime_rate_cat.phone.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.website.fillna(DC_crime_rate_cat.website.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.intake1.fillna(DC_crime_rate_cat.intake1.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.intake2.fillna(DC_crime_rate_cat.intake2.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.Geography.fillna(DC_crime_rate_cat.Geography.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.Geographic_Area_Name.fillna(DC_crime_rate_cat.Geographic_Area_Name.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.ANC.fillna(DC_crime_rate_cat.ANC.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.BID.fillna(DC_crime_rate_cat.BID.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.END_DATE.fillna(DC_crime_rate_cat.END_DATE.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.VOTING_PRECINCT.fillna(DC_crime_rate_cat.VOTING_PRECINCT.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.BLOCK_GROUP.fillna(DC_crime_rate_cat.BLOCK_GROUP.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.START_DATE.fillna(DC_crime_rate_cat.START_DATE.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.sector.fillna(DC_crime_rate_cat.sector.value_counts().idxmax(), inplace=True)
DC_crime_rate_cat.NEIGHBORHOOD_CLUSTER.fillna(DC_crime_rate_cat.NEIGHBORHOOD_CLUSTER.value_counts().idxmax(), inplace=True)
```

```python
DC_crime_rate_cat=DC_crime_rate_cat.apply(le.fit_transform)
```
